```python
df = df.drop(columns=['Mocodes'])

df = df.dropna(subset=['Crm Cd 1'])

df['Vict Sex'] = df['Vict Sex'].fillna('Unknown')
df['Vict Descent'] = df['Vict Descent'].fillna('Unknown')

df['Weapon Used Cd'] = df['Weapon Used Cd'].fillna(0)
df['Weapon Desc'] = df['Weapon Desc'].fillna('No Weapon')

df = df.drop(columns=['Premis Cd'])

df['Premis Desc'] = df['Premis Desc'].fillna('Unknown')

df = df.drop(columns=['Crm Cd 2', 'Crm Cd 3', 'Crm Cd 4'])

df = df.drop(columns=['Cross Street'])

df['Vict Sex'] = df['Vict Sex'].replace({'X': 'Unknown'})

df.isnull().sum()
```

-There are multiple columns with large amounts of null values. "Mocodes" describes how a crime was committed, such as if there was a weapon involved or if the victim is a family member. I believe this column is not necessary and should be dropped. It is messy and difficult to analyze, and we are more focused on what the crime is rather than how it was carried out. <br>-"Victim sex" and "descent" have a large amount of null values but they are necessary to understand the demographic information of this dataset. The values that are null will be filled with unknown to ensure they are accounted for. <br>-The column " Premis Desc" will be filled with "Unknown" to account for the missing values. "Premis Cd" is not necessary and will be dropped.<br>-There are two categories containing weapon info, "Weapon Used Cd" and "Weapon Desc." The null values will be filled with "No Weapon" to account for the crimes without weapons involved. <br>-The 4 columns named "Crm Cd" describe if a suspect has multiple charges. "Crm Cd 1" is the primary charge and any after that are secondary. There are 11 null values under "Crm Cd 1." These null values can be dropped from the dataset as they cannot be categorized. The remaining columns showing the other charges can all be dropped as well. The amount of null values is too high and they are not essential to understanding the dataset.<br>-Lastly, the column "Cross Street" is not essential to the dataset and will be dropped. The location and cordinates are already provided.

These are the LAPD case status codes. In order to understand the data it is essential to understand what each of these means.<br>
IC- Investigation Continuing (still active and open case)<br>
AO- Adult Other (adult was involved but not arrested)<br>
AA- Adult Arrest<br>
JA- Juvenile Other (juvenile was involved but not arrested)<br>
JO- Juvenile Arrest <br>
CC- Cleared by Other Means<br>

```python
df['Date Rptd'] = pd.to_datetime(df['Date Rptd'])
df['DATE OCC'] = pd.to_datetime(df['DATE OCC'])

df['Year'] = df['DATE OCC'].dt.year
df['Month'] = df['DATE OCC'].dt.month
df['Month Name'] = df['DATE OCC'].dt.strftime('%B')
df['Day of Week'] = df['DATE OCC'].dt.strftime('%A')

df['Hour'] = df['TIME OCC'].astype(str).str.zfill(4).str[:2].astype(int)
```

This code makes the data easier to understand and visualize. It reformats the dates when the crimes occured and were reported into a simpler format that reads YEAR-MONTH-DAY. It also creates five new columns "Year," "Month," "Month Name," "Day of the Week," and "Hour." These will make it easier to visualize and categorize the crimes.

```python
descent_map = {
    'A': 'Other Asian',
    'B': 'Black',
    'C': 'Chinese',
    'D': 'Cambodian',
    'F': 'Filipino',
    'G': 'Guamanian',
    'H': 'Hispanic/Latin/Mexican',
    'I': 'American Indian/Alaskan Native',
    'J': 'Japanese',
    'K': 'Korean',
    'L': 'Laotian',
    'O': 'Other',
    'P': 'Pacific Islander',
    'S': 'Samoan',
    'U': 'Hawaiian',
    'V': 'Vietnamese',
    'W': 'White',
    'X': 'Unknown',
    'Z': 'Asian Indian'
}

df['Vict Descent'] = df['Vict Descent'].map(descent_map)
```

```python
df['Vict Descent'] = df['Vict Descent'].fillna('Unknown')
```
